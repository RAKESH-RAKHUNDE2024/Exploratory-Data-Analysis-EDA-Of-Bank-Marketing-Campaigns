### 7) Correlation Analysis
#### Question 7) :- Compute The Correlation Matrix For Numerical Columns And Visualize It Using A Heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns
numerical_df = Banking_Df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numerical_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

### 8) Outlier Detection
#### Question 8) :- Detect Outliers In Numerical Columns Using The Interquartile Range (IQR) Method.

```python
# Define A Function To Detect Outliers

def detect_outliers(Banking_Df, column):
    Q1 = Banking_Df[column].quantile(0.25)
    Q3 = Banking_Df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return Banking_Df[(Banking_Df[column] < lower_bound) | (Banking_Df[column] > upper_bound)]

# Detect Outliers In 'duration' Column
outliers_duration = detect_outliers(Banking_Df, 'duration')
outliers_duration
```

### 10) Bivariate Analysis
#### Question 10) :- Analyze The Relationship Between 'duration' And 'y' (Subscription To Term Deposit) Using A Box Plot.

```python
# Box Plot Of 'duration' By 'y'

plt.figure(figsize=(8,6))
sns.boxplot(x='y', y='duration',data=Banking_Df, color ='red')
plt.title('Duration Vs Subscription')
plt.show()
```

### 11) Pivot Tables
#### Question 11) :- Create A Pivot Table To Analyze The Average 'duration' Of Calls For Each 'job' Category.

```python
# Create Pivot Table

pivot_table = Banking_Df.pivot_table(values='duration',index='job',aggfunc='mean')
pivot_table.sort_values(by='duration',ascending=False)
```

### 12) Subscription Rate By Job Type
#### Question 12) :- Calculate The Subscription Rate (Percentage Of 'yes' In 'y') For Each 'job' Category.

```python
# Calculate Subscription Rate By Job

subscription_rate = Banking_Df.groupby('job')['y'].value_counts(normalize=True).unstack().iloc[:,1] * 100
subscription_rate.sort_values(ascending=False)
```

### 13) Age Distribution By Subscription Status
#### Question 13) :- Visualize The Distribution Of 'age' For Clients Who Subscribed ('y' = 'yes') And Did Not Subscribe ('y' = 'no') Using Histograms.

```python
# Plot Age Distribution By Subscription Status

plt.figure(figsize=(10,6))
sns.histplot(Banking_Df[Banking_Df['y'] == 'yes']['age'], color='green', label='Subscribed', kde=True, stat='density')
sns.histplot(Banking_Df[Banking_Df['y'] == 'no']['age'], color='orange', label='Not Subscribed',kde=True, stat='density')
plt.title('Age Distribution By Subscription Status')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

### 14) Call Duration Analysis
#### Question 14) :- Determine The Average Call Duration For Each 'contact' Type.

```python
# Calculate Average Duration By Contact Type

avg_duration_contact =Banking_Df.groupby('contact')['duration'].mean()
avg_duration_contact.sort_values(ascending=False)
```
